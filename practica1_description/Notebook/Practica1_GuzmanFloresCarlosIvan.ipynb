### Definicin de URDF

El URDF (Unified Robot Description Format) es un archivo en formato XML diseado para describir de manera detallada la estructura cinemtica de un robot, sus parmetros dinmicos, la representacin visual y las geometras de colisin. Este archivo puede incluir enlaces a otros archivos que contienen modelos 3D de las distintas partes del robot, formando lo que se conoce como un "URDF Bundle". El formato URDF fue creado para ser un modelo autnomo que agrupa toda la informacin necesaria para describir un robot en un solo archivo, lo que facilita su integracin a travs de herramientas de anlisis XML. Aunque su origen est en ROS, su independencia ha permitido que sea adoptado por mltiples herramientas, tanto dentro como fuera del entorno de ROS [1].

### Tipos de juntas que estn definidas dentro del formato URDF

Los tipos de juntas soportados en URDF son:

- **Revoluta**: Permite rotacin limitada.

- **Continua**: Similar a la revoluta, pero sin lmites en la rotacin.

- **Prismtica**: Permite movimiento lineal a lo largo de un eje.

- **Fija**: Conecta dos enlaces sin permitir movimiento relativo entre ellos.

- **Flotante**: Permite seis grados de libertad (tres de rotacin y tres de traslacin).

- **Plana**: Permite deslizamiento en un plano.


### Definicin de elementos "primitivos" que se pueden utilizar para definir un robot en ROS

Los elementos primitivos en URDF incluyen enlaces (links) y juntas (joints). Los enlaces representan cuerpos rgidos, mientras que las juntas conectan dos enlaces permitiendo el movimiento relativo entre ellos. Adems, se utilizan geometras simples como cajas y cilindros para definir las propiedades visuales y de colisin de los enlaces.

### Qu es un nodo y que es un topico en ROS 2

- **Nodo**: En ROS 2, un nodo es una unidad de proceso que puede realizar tareas especficas. Los nodos pueden comunicarse entre s mediante el envo y recepcin de mensajes.

- **Tpico**: Es un canal de comunicacin utilizado por los nodos para intercambiar mensajes de forma asincrnica. Un nodo puede publicar informacin en un tpico, mientras que otros nodos pueden suscribirse para recibir esos datos.

### Qu es RVIZ

RVIZ es una herramienta de visualizacin en ROS utilizada para representar de manera grfica los robots y el entorno en el que operan. Soporta la visualizacin de modelos URDF, proporcionando una interfaz para inspeccionar tanto las geometras visuales como de colisin de un robot.
